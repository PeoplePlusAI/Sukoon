```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader=PyPDFLoader('FINAL-Immediate Care.pdf')
docs=loader.load()
```

```python
OPENAI_API_KEY = ""
```

```python
from langchain_objectbox.vectorstores import ObjectBox ##vector Database
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(docs)
```

```python
from langchain_openai import OpenAIEmbeddings
vector = ObjectBox.from_documents(documents, OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY), embedding_dimensions=768)
vector
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains import RetrievalQA
from langchain import hub
```

```python
llm = ChatOpenAI(model="gpt-4o", openai_api_key=OPENAI_API_KEY) ## Calling Gpt-4o
prompt = hub.pull("rlm/rag-prompt")
prompt
```

```python
qa_chain = RetrievalQA.from_chain_type(
        llm,
        retriever=vector.as_retriever(),
        chain_type_kwargs={"prompt": prompt}
    )
```

```python
# question = "What is immediate care"
question = "Who can provide immediate care"
result = qa_chain({"query": question })
result
```

# Hugging Face Models

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS

from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain.prompts import PromptTemplate

from langchain.chains import RetrievalQA
import  numpy as np
```

```python
loader=PyPDFLoader("FINAL-Immediate Care.pdf")

documents=loader.load()

text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)

final_documents=text_splitter.split_documents(documents)
final_documents[0]
```

```python
len(final_documents)
```

```python
huggingface_embeddings=HuggingFaceBgeEmbeddings(
    model_name="BAAI/bge-small-en-v1.5",      #sentence-transformers/all-MiniLM-l6-v2
    model_kwargs={'device':'cpu'},
    encode_kwargs={'normalize_embeddings':True}

)
```

```python
print(np.array(huggingface_embeddings.embed_query(final_documents[0].page_content)))
print(np.array(huggingface_embeddings.embed_query(final_documents[0].page_content)).shape)
```

```python
vectorstore=FAISS.from_documents(final_documents,huggingface_embeddings)
```

```python
retriever=vectorstore.as_retriever(search_type="similarity",search_kwargs={"k":3})
print(retriever)
```

```python
huggingface_access_token = ""
import os
os.environ['HUGGINGFACEHUB_API_TOKEN']=huggingface_access_token
```

```python
from langchain_community.llms import HuggingFaceHub

hf=HuggingFaceHub(
    repo_id="mistralai/Mistral-7B-v0.1",
    model_kwargs={"temperature":0.01,"max_length":500},
    huggingfacehub_api_token=huggingface_access_token
)
```

```python
query="Who can provide immediate care?"
hf.invoke(query)
```

```python
!huggingface-cli login --token $HUGGINGFACEHUB_API_TOKEN --add-to-git-credential
```

```python
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline

hf = HuggingFacePipeline.from_model_id(
    model_id="mistralai/Mistral-7B-v0.1",
    task="text-generation",
    pipeline_kwargs={"temperature": 0, "max_new_tokens": 300}
)

llm = hf
llm.invoke(query)
```

```python
prompt_template="""
Use the following piece of context to answer the question asked.
Please try to provide the answer only based on the context

{context}
Question:{question}

Helpful Answers:
 """

prompt=PromptTemplate(template=prompt_template,input_variables=["context","question"])
retrievalQA=RetrievalQA.from_chain_type(
    llm=hf,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={"prompt":prompt}
)
# query="""DIFFERENCES IN THE
# UNINSURED RATE BY STATE
# IN 2022"""

result = retrievalQA.invoke({"query": query})
print(result['result'])
```
